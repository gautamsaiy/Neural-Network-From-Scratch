```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sys
sys.path.append("..")
from utils import *
```

```python
plt.style.use("seaborn")
input = np.arange(-5, 5.1, .01)
```

```python
output_ReLU = forward_ReLU(input)
output_deriv_ReLU = deriv_ReLU(input)
sns.lineplot(x=input, y=output_ReLU, label="ReLU")
sns.lineplot(x=input, y=output_deriv_ReLU, label="Derivative", linestyle="--")
plt.title("ReLU and the Derivative ReLU")
```

```python
output_sigmoid = forward_sigmoid(input)
output_deriv_sigmoid = deriv_sigmoid(input)
sns.lineplot(x=input, y=output_sigmoid, label="Sigmoid")
sns.lineplot(x=input, y=output_deriv_sigmoid, label="Derivative", linestyle="--")
plt.title("Sigmoid and its Derivative")
```
